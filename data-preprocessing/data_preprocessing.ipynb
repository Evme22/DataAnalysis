Проверка датасета на наличие пропусков с помощью метода isna(), подсчет пропусков с помощью метода sum().

Во всех колонках, кроме "duration" и "main_genre", есть пропуски. Заполнение пропусков. Для колонки "title" в качестве заменителя записывается – "Unknown film". Для колонки "main_production" – "Unknown production". Для числовых колонок заменителем будет выставлено среднее значение. Првторный вывод сводки по количеству пропусков в каждом столбце.

```python
df['title'] = df['title'].fillna("Unknown film")
df['main_production'] = df['main_production'].fillna("Unknown production")
df['release_year'] = df['release_year'].fillna(df['release_year'].mean())
df['score'] = df['score'].fillna(df['score'].mean())
df['number_of_votes'] = df['number_of_votes'].fillna(df['number_of_votes'].mean())
df.isna().sum()
```

Все пропуски заполнены. \
Поиск явных дубликатов в данных, подсчет их количества с помощью методов .duplicated() и .sum().

```python
print(df.duplicated().sum())
```

Удаление дубликатов (drop_duplicates()) и обновление индексации (reset_index(drop=True)), чтобы избавиться от пропусков. Аргумент drop=True используется для того, чтобы при обновлении индексации не создался новый столбец с индексами.

```python
df = df.drop_duplicates().reset_index(drop=True)
```

Проверка типов данных, вывод информацию о датасете при помощи метода info(). Приведение столбца release_year к числовому типу данных.

```python
df = df.astype({"release_year" : 'int64'})
print(df.info())
```

Для столбцов с числовыми данными определены числовые типы данных, для столбцов с текстовыми данными - object. \
Группировка по жанру и количеству фильмов каждого main_production. \
Для выполнения используются методы groupby() и agg().


```python
grouped_df = df.groupby(['main_genre', 'main_production']).agg({'main_production': ['count']})
grouped_df.columns = ['count']
grouped_df
```

Сортировка датафрейма по убыванию столбца 'count' с помощью метода sort_values(). Вывод отсортированного датафрейма.

```python
sorted_df = grouped_df.sort_values(by='count', ascending=False)
sorted_df
```

Больше всего фильмов снято в жанре drama. Больше всего фильмов в жанре drama снято производством US – 55. Меньшее количество снятых фильмов встречается в разных жанрах и производствах фильмов. \
Создается сводная таблица по средниему рейтингу фильмов по main_production. Сортировка по убыванию рейтинга.

```python
data_pivot = pd.pivot_table(df, index='main_production', values='score', aggfunc='mean')
sorted_data_pivot = data_pivot.round(1).sort_values(by='score', ascending=False)
sorted_data_pivot
```

Как видно из результатов сортировки сводной таблицы, самый большой средний рейтинг у кинопроизводства CD (8,2), а самый маленький у ID (6,9). \
Создается сводная таблица по медианному рейтингу фильмов по жанрам и main_production. Сортировка по возрастанию (от A до Z) main_production.

```python
data_pivot = pd.pivot_table(df, index='main_production', columns='main_genre', values='score', aggfunc='median')
sorted_data_pivot = data_pivot.sort_values(by='main_production', ascending=True)
sorted_data_pivot
```

Исходя из данных полученной таблицы можно увидеть медианные оценки фильмов в различных жанрах всех производств, отсортированных в порядке возрастания по производству. Например, медианная оценка производства фильмов AR в жанре comedy составляет 7.5. Медианная оценка производства фильмов KR в жанре triller составляет 7.25.


Выделяется топ 7 жанров с самым высоким рейтингом. Так как к одному жанру может принадлежать несколько фильмов, соответственно, несколько оценок, рассматриваются все жанры и  средний рейтинг по каждому жанру. Используется группировка данных по жанрам. Для каждого жанра вычисляется средний рейтинг с помощью .mean().

```python
mean_score = df.groupby('main_genre')['score'].mean()
mean_score
```

Сортировка в порядке убывания полученного рейтинга жанров. Срез по индексу с помощью .index[:7]. Результат top_genres - список топ 7 жанров с самым высоким средним рейтингом.

```python
mean_score_sorted = mean_score.round(2).sort_values(ascending=False)
top_genres = mean_score_sorted.index[:7]
top_genres
```

Фильтрация датафрейма в соответсвии с полученным топом жанров. Результат фильтрации - датафрейм, содержащий фильмы по 7 жанрам.

```python
filtered_df = df[df['main_genre'].isin(top_genres)]
filtered_df
```

Выборка топ 10 лет по количеству записей. Для этого используется функция .value_counts(). Результат - отсортированный массив с перечнем годов и соответсвующим количеством фильмов.

```python
year_counts = filtered_df['release_year'].value_counts()
year_counts
```

Срез первых 10 лет, где количество фильмов максимальное. Результат - список топ 10 лет по количеству фильмов.

```python
top_years = year_counts.index[:10]
top_years
```

Фильтрация датафрейма в соответсвии с полученным топом жанров. Результат фильтрации - датафрейм с фильмами из топ 7 жанров и топ 10 лет.

```python
filtered_df_by_year_and_genre = filtered_df[filtered_df['release_year'].isin(top_years)]
filtered_df_by_year_and_genre
```

Создание сводной таблицы. В качестве аргументов сводной таблицы передается 'release_year' в строки (выборка топ 10 лет), 'main_genre' в столбцы (выборка топ 7 жанров), 'duration' в значение.

```python
pivot_table = pd.pivot_table(filtered_df_by_year_and_genre, index='release_year', columns='main_genre', values='duration', aggfunc='median')
pivot_table
```

Эта сводная таблица представляет медианное значение длительности фильмов в минутах для разных жанров и лет (с 2010 по 2021 годы) по семи жанрам: action, comedy, documentary, drama, scifi, war и western. Для некоторых жанров и лет отсутствуют данные. Например, для жанров action, documentary, scifi, war и western часто встречаются пропуски, вследствии недостаточного количества фильмов этих жанров в рассматриваемых годах для расчета медианы. Жанры comedy и drama присутствуют почти во всех годах. Это может указывать на их высокую популярность и стабильное производство фильмов этих жанров.
Сводная таблица показывает, что более популярные жанры, такие как comedy и drama, имеют данные по большинству лет, в то время как жанры, такие как action, scifi, war и western, представлены с пропусками. Некоторые жанры (например, scifi, western) часто имеют высокие медианные значения, что свидетельствует о том, что они склонны к созданию более длинных фильмов.
